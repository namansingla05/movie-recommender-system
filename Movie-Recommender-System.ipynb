```python

def convrt(obj):
    L = []
    for i in obj:
        L.append(i['name'])
    return L

#but this will not work beacause this is string of list = '[]'
#so we need to convert the string of list into a list
#for this we have module ast and in that we have a s=function of name literal_eval()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#now for cast the first dictionary conatains the most important start cast and so on... but we want only 4 top 
#and we want only name of that 4 top cast

def convertcast(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 4:
            L.append(i['name'])
            count += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convertcast)
```

```python
movies['cast']
```

```python
#now in this crew we want only director ans one cell contains many dictionary for differnt roles but we want only where job role is director
movies['crew'][0]
```

```python
def convertcrew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(convertcrew)
```

```python
#now overview is a string so we need to convert this into list 
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview']
```

```python
#now the thing is we need to remove the spaces between the words like sam martin so sam and martin behaves as differnt tag
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
#now as we know all the columns like overview cast keywords crew and genres behave as tags so we can merge that
movies['tags']= movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('loved')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out().tolist()
```

```python
#now as we need to find the distance between all the movcies or vector but data is of highefr dimension so we not use the 
#eucledian distance we find cosine distance(angle between the vector)

from sklearn.metrics.pairwise import cosine_similarity 
#tow fuction one for similarity and one for distance so we need similarity and similarity is inversionaly proportional to distnace
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse = True , key= lambda x:x[1])
```

```python
#the thing is we can sort the distances but then their is problem the index losses(we cant find the distance for specific index)
#so we need to keep the index also for this we use enumerate function
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    distances  = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True , key= lambda x:x[1])[1:6] #want first five 

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```

```python
recommend('Battle: Los Angeles')
```
